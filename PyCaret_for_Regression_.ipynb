# **2. Regression: Advance - 1**

#### **2.1 Model Performance using data "Normalization"**

```python
s = setup(data = bostonDataSet, target = 'medv', normalize = True, normalize_method = 'zscore')
cm = compare_models()

#normalize_method = {zscore, minmax, maxabs, robust}
```

---
### **2.2 Model Performance using "Feature Selection"**
---

```python
s = setup(data = bostonDataSet, target = 'medv', feature_selection = True)
cm = compare_models()
```

---
### **2.3 Model Performance using "Outlier Removal"**
---

```python
s = setup(data = bostonDataSet, target = 'medv', remove_outliers = True, outliers_threshold = 0.05)
cm = compare_models()
```

---
### **2.4 Model Performance using "Transformation"**
---

```python
s = setup(data = bostonDataSet, target = 'medv', transformation = True, transformation_method = 'yeo-johnson')
cm = compare_models()
```

---
### **2.5 Model Performance using "PCA"**
---

```python
s = setup(data = bostonDataSet, target = 'medv', pca = True, pca_method = 'linear')
cm = compare_models()
```

---
### **2.6 Model Performance using "Outlier Removal" + "Normalization"**

```python
s = setup(data = bostonDataSet, target = 'medv', remove_outliers = True, outliers_threshold = 0.05, normalize = True, normalize_method = 'zscore')
cm = compare_models()
```
